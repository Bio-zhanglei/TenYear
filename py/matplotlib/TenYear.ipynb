```python
%matplotlib inline
font = {'font.sans-serif': 'MicroSoft YaHei','axes.unicode_minus': False}
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file1 = "SCI_year.tsv"
# file2 = "SCI_style.tsv"
file3 = "Uni_year.tsv"

#修改label,tick,title的一个自定义函数
def line_annotation(xList, yList, title='????', xLabel=None, yLabel=None, xTickRotation=45,fontsize=8):
    plt.title(title)#标题
    plt.xlabel(xLabel)#xlabel
    plt.ylabel(yLabel)#ylabel
    plt.xticks(rotation=45)#xtick旋转角度
    for x, y in zip(xList, yList):
        plt.text(x, y+0.5, str(y), ha='center', va='bottom', fontsize=fontsize)#数据标签
```

```python
data = pd.read_csv(file1, sep='\t',encoding="GBK",index_col=0)
data.T.to_json('SCI_year.json',force_ascii=False)
```

```python
#透明背景色
data = pd.read_csv(file3, sep='\t',encoding="GBK",dtype={0:str})
data.drop(['河南牧业经济学院','吉林农业科技学院','信阳农林学院','仲恺农业工程学院'],axis=1,inplace=True)

for j in range(9):
    fig = plt.figure(dpi=100,figsize=(8,6),num=j)
    sns.set(style='ticks',rc=font, font_scale=1.2)
    
    fig.subplots_adjust(wspace=0.4,hspace=0.8)
    fig.patch.set_alpha(0)
    if (j %2 ==0):
        color = ['#bc002d','#002a8f','#bc002d','#002a8f']
    else:
        color = ['#002a8f','#bc002d','#002a8f','#bc002d']
    flag=0
    for i in range(4*j,4*(j+1),2):
#         fig.patch.set_facecolor('#eff5f5')
        ax = fig.add_subplot(2,2,flag+1)
#         ax.patch.set_facecolor('#eff5f5')
        ax.patch.set_alpha(0)
        sns.lineplot(x = data.iloc[:,0],
                 y = data.columns.to_list()[i+1],
                 data = data,
                 marker='o',
                 markersize=4,
                 color =color[flag],
                 ax=ax
                 )
        sns.despine()
        plt.ylim(0,5000)
        line_annotation(data.iloc[:,0], data.iloc[:,i+1], title=data.columns.to_list()[i+1])
        ax = fig.add_subplot(2,2,flag+2)
#         ax.patch.set_facecolor('#eff5f5')
        ax.patch.set_alpha(0)
        sns.lineplot(x = data.iloc[:,0],
                 y = data.columns.to_list()[i+2],
                 data = data,
                 marker='o',
                 markersize=4,
                 color =color[flag],
                 ax=ax
                 )
        sns.despine()
        plt.ylim(0,5000)
        line_annotation(data.iloc[:,0], data.iloc[:,i+2], title=data.columns.to_list()[i+2])
        flag += 2
    plt.savefig("Uni_SCI-{0}.png".format(j+1))
#     如果设置了背景色，要设置savefig是要设置facecolor属性，不然还是默认的白色
#     plt.savefig("Uni_SCI-{0}.png".format(j+1), facecolor=fig.get_facecolor())
```

```python
#底色
data = pd.read_csv(file3, sep='\t',encoding="GBK",dtype={0:str})
data.drop(['河南牧业经济学院','吉林农业科技学院','信阳农林学院','仲恺农业工程学院'],axis=1,inplace=True)

for j in range(9):
    fig = plt.figure(dpi=100,figsize=(8,6),num=j)
    sns.set(style='ticks',rc=font, font_scale=1.2)
    
    fig.subplots_adjust(wspace=0.4,hspace=0.8)
    if (j %2 ==0):
        color = ['#bc002d','#002a8f','#bc002d','#002a8f']
    else:
        color = ['#002a8f','#bc002d','#002a8f','#bc002d']
    flag=0
    for i in range(4*j,4*(j+1),2):
        ax = fig.add_subplot(2,2,flag+1)
        sns.lineplot(x = data.iloc[:,0],
                 y = data.columns.to_list()[i+1],
                 data = data,
                 marker='o',
                 markersize=4,
                 color =color[flag],
                 ax=ax
                 )
        sns.despine()
        plt.ylim(0,5000)
        line_annotation(data.iloc[:,0], data.iloc[:,i+1], title=data.columns.to_list()[i+1])
        ax = fig.add_subplot(2,2,flag+2)
        sns.lineplot(x = data.iloc[:,0],
                 y = data.columns.to_list()[i+2],
                 data = data,
                 marker='o',
                 markersize=4,
                 color =color[flag],
                 ax=ax
                 )
        sns.despine()
        plt.ylim(0,5000)
        line_annotation(data.iloc[:,0], data.iloc[:,i+2], title=data.columns.to_list()[i+2])
        flag += 2
    plt.savefig("Big-Uni_SCI-{0}.png".format(j+1), facecolor=fig.get_facecolor())
```
